```python
import numpy as np 
import matplotlib.pyplot as plt
#plt.ion()
%pylab inline
```

```python
x = np.linspace(0,30,2000)
y = np.sin(x) + 0.8*(np.random.random(len(x)) - 0.5)
figure(figsize=(14,8))
plot(x,y)
```

```python
def smooth(x,y,nnb):
    def lo_lim(i):
        return i if i > 0 else 0
    def hi_lim(i):
        return i+1 if i < len(x)-1 else len(x)
    
    yy = np.zeros(len(x))
    
    for i in np.arange(len(x)):
        msk = np.arange(lo_lim(i-nnb),hi_lim(i+nnb))
        yy[i] = np.mean(y[msk])
    
    return yy
```

```python
figure(figsize=(14,8))
plot(x,smooth(x,y,30))
```
