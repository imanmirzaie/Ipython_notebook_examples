```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(0,10,500)
y1 = np.sinc(x)
```

```python
fig, ax = plt.subplots()

# figure properties
fig.set_size_inches((12,8))

# axes (not axis) properties

# axes line
ax.plot(x,y1,linewidth=3) # add data


# axes title
ax.set_title('Axes title')
ax.title.set_position((0.5,1.02))
ax.title.set_fontsize(26)
ax.title.set_fontname('Times New Roman')

# axis properties
ax.axhline(linewidth=2,color='r') # extra horizontal line

for sp in ax.spines.values(): # spine properties
    sp.set_linewidth(2)
    

# axis ticks and tick labels
ax.tick_params(axis='both', length=6, width=2, pad=10)
tickfonts = matplotlib.font_manager.FontProperties(family='Times New Roman',size=22)
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontproperties(tickfonts)

# axis labels
ax.xaxis.set_label_text('x-axis label',fontdict={'fontname':'Times New Roman',
                                      'fontsize':24});
ax.yaxis.set_label_text('y-axis label',fontdict={'fontname':'Times New Roman',
                                      'fontsize':24});
```
